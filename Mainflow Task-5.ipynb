```python
df.AHD.value_counts()
```

```python
df.AHD.value_counts().plot(kind='bar',color=["orchid","salmon"])
plt.title("Heart Disease Values")
plt.xlabel("1=Heart Disease,0=No Heart Disease")
plt.ylabel("Amount")
```

```python
df.AHD.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Disease','No Disease'])
```

```python
df.Sex.value_counts()
```

```python
df.Sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend('Male Female Ratio')
plt.legend(['Male','Female'])
```

```python
pd.crosstab(df.AHD,df.Sex)
```

```python
sns.countplot(x='AHD',data=df,hue='Sex')
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('No=No Heart Disease,Yes=Heart Disease')
```

```python
df.Ca.value_counts()
```

```python
df.Ca.value_counts().plot(kind='bar',color=["salmon","lightskyblue","springgreen","khaki"])
plt.title("Chest Pain type vs count")
```

```python
pd.crosstab(df.Sex,df.Ca)
```

```python
pd.crosstab(df.Sex,df.Ca).plot(kind='bar',color=['coral','lightskyblue','plum','khaki'])
plt.title("Type of chest pain for sex")
plt.xlabel('0=Female,1=Male')
```

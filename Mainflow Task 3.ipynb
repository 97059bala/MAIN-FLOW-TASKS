```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv("Diabetes Missing Data.csv")
```

```python
data.head()
```

```python
#scatter plot between pregnant vs Glucose
plt.scatter(data['Pregnant'],data['Glucose'])
#Title of Scatter Plot
plt.title("Scatter Plot")
#Label the x and y
plt.xlabel('Pregnant')
plt.ylabel('Glucose')
plt.show()
```

```python
#Line Chart against Pregnant vs Glucose
plt.plot(data['Pregnant'])
plt.plot(data['Glucose'])
#Title
plt.title("Line chart")
#Label the x and y
plt.xlabel('Pregnant')
plt.ylabel('Glucose')
plt.show()
```

```python
#Bar Chart or Bar Graph
plt.bar(data['Pregnant'],data['Glucose'])
#Title
plt.title("Bar Chart")
#Label the x and y
plt.xlabel('Pregnant')
plt.ylabel('Glucose')
plt.show()
```

```python
#Histogram
plt.hist(data['Pregnant'])
plt.title("Histogram")
plt.show()
```
